Un problema común en Ingeniería Civil corresponde al diseño de las cimentaciones de los edificios. Estas son necesarias toda vez que la mayoría de tipos de suelos de soporte no tienen la resistencia mecánica suficiente para transmitir de manera continua y funcional las cargas del edificio. En la práctica industrial es común resolver el problema de diseño mediante un procedimento de 2 pasos en el que generalmente participan profesionales en Geotecnia y en Diseño Estructural. En un primer paso el Ingeniero Geotecnista, utilizando métodos de campo y ensayos de laboratorio, determina las propiedades mecánicas del depósito de suelo (ver tabla adjunta). A la vez éste Ingeniero Geotecnista suele hacer recomendaciones acerca del tipo de cimentación usando estas propiedades. Estas generalmente están asociadas a las cargas transmitidas por el edificio.

Dependiendo del tipo de recomendación el geotecnista entrega también al diseñador estructural las resistencias requeridas, generalmente en términos de rigideces efectivas (de resortes equivalentes) para ser usadas posteriormente en el diseño estructural de la cimentación utilizando modelos simplificados. Por ejemplo es común que el Geotecnista suminstre la tensión admisible y rigidez efectivas para el diseño de zapatas así como la tensión admisible (resistencia por punta), resistencia a la fricción y rigideces laterales para el diseño de cimentaciones profundas.

## Condiciones de diseño de la sub-estructura

Los sistemas de cimentación que podrán ser usados son los siguientes:

* (i) Zapatas cuadradas de ancho $B$.

* (ii) Pilotes cilíndricos de diámetro (D) de $50$ cm, $70$ cm y $90$ cm. Los pilotes podrán actuar de manera individual o en grupo.


### Propiedades mecánicas del concreto
 
Como material base para el concreto de la cimentación use las siguientes propiedades:

* Módulo de elasticidad: $2.0 \, Mtf/m^2$.

* Relación de Poisson: $0.20$.

* Resistencia a la compresión: $2100 \, tf/m^2$.

* Resistencia a la tensión: $2100 \, tf/m^2$(Este valor no es real y solo es asumido para fines académicos).

* Resistencia al corte : $1000 \, tf/m^2$ (Este valor no es real y solo es asumido para fines académicos).

### Flujo de trabajo en un análisis por elementos finitos en SolidsPy.

Un flujo de trabajo típico en `SolidsPy` requiere:

* Creación de la malla en `gmsh`.

* Creación de archivos de texto usando un script de mallado.

* Ejecución del análisis por elementos finitos.

La carpeta `files` contiene los modelos necesarios para la actividad. Cada modelo esta conformado por un archivo de extensión `.geo` en el cual se encuentran definidos los paramétros geometricos del modelo como espesores de los estratos, tamaño de la zapata, ancho de la columna, etc. Estos valores pueden ser modificados con cualquier editor de texto.

Para crear la malla abra el archivo `.geo` desde el programa `gmsh` y seleccione la ruta `>mesh>define>2d`. Esto producirá un archivo con extensión `.msh` el cual contiene toda la información de la malla.
   
### Algunas funciones de post-procesado de SolidsPy

* `fields_plot()` : Grafica los contornos de los campos tensoriales de tensiones y deformaciones resultantes del análisis.

* `principal_dirs()` : Calcula los valores y direcciones principales de un tensor simétrico de orden 2.

* `plot_node_field()` : Grafica contornos para cualquier campo especificado en los nodos de la malla.

* `filtered()` : Resalta las zonas del modelo donde se excede un valor de referencia de un campo como por ejemplo la resistencia de un material.


     
### Algunas unidades y equivalencias

* $1 \quad \text{ N} = 1\quad \text{ kg m/s}^2$.

* $1\quad \text{ Pa} = 1 \text{ N/m}^2$.

* $1\quad \text{ kPa} = 1 \times 10^3 \text{ Pa}$.

* $1\quad \text{ MPa} = 1 \times 10^6 \text{ Pa}$.

* $1\quad \text{ GPa} = 1 \times 10^9 \text{ Pa}$.

* $1 \quad \text{ MPa} = 100 tf/m^2$.
